# *User Input*
User input sangat penting untuk membangun program interaktif. Setiap bahasa pemrograman memiliki implementasi masing-masing untuk `input`. C++ memiliki `scanf`, Java memiliki kelas `Scanner`, dan Ruby memiliki `get`.

Meskipun dalam kebanyakan kasus, input ini berasal dari keyboard, bentuk lain seperti klik mouse, track-pad, sensor, dll. juga mungkin digunakan.

Seperti biasa, Python menyediakan kerangka kerja sederhana untuk mendapatkan input pengguna dalam bentuk fungsi `input()`.
Fungsi `input()` membaca baris dari konsol, mengubahnya menjadi **string**, dan mengembalikannya.

Syntax
`input([prompt])`

**prompt** adalah argumen string opsional, digunakan untuk menampilkan pesan bagi pengguna. Contoh: 'Masukkan Nama:'
Ketika fungsi `input()` dipanggil, aliran program berhenti sampai pengguna memasukkan beberapa input. Pengguna kemudian menambahkan beberapa informasi dan menekan tombol Enter. Fungsi `input()` kembali ke program dengan **string** yang dimasukkan.

Contoh:

```python
Variabel_input = input('Masukkan nama : ')
print(Variabel_input)
```

## *Input Type*
Fungsi `input()` mengubah semua informasi yang diterimanya ke dalam format **string**.
Mari kita lihat:

```python
name = input("Enter name: ")
age = input("Enter age: ")

print("data_type of name: ", type(name))
print("data_type of age: ", type(age))
```

Kita menggunakan fungsi `type()` untuk memeriksa tipe data objek. Seperti yang Anda lihat di atas, tidak masalah apakah kita memasukkan angka atau string karakter. Keduanya kembali sebagai objek **string**.

## Merubah jenis data input
Karena fungsi `input()` mengembalikan semuanya sebagai **string**, kita perlu melakukan beberapa konversi tipe eksplisit (konversi yang dilakukan manual oleh user) input menjadi jenis data yang dikehendaki.

Beberapa *function* yang bisa digunakan untuk konversi data di Python yaitu:
1. `int(a, base)`: Fungsi ini mengubah semua tipe data menjadi **integer**. 'Base' menentukan basis di mana string berada jika tipe datanya adalah string.
2. `float()`: Fungsi ini digunakan untuk mengubah tipe data apa pun menjadi bilangan **floating-point**.
3. `ord()` : Fungsi ini digunakan untuk mengubah karakter menjadi teks **unicode** yang sesuai.
4. `hex()` : Fungsi ini untuk mengubah string integer menjadi **heksadesimal**.
5. `oct()` : Fungsi ini untuk mengubah string integer menjadi **octal**.
6. `tuple()` : Fungsi ini digunakan untuk mengkonversi ke **tuple**.
7. `set()` : Fungsi ini mengembalikan tipe setelah dikonversi ke **set**.
8. `list()` : Fungsi ini digunakan untuk mengonversi tipe data apa pun menjadi tipe **list88.
9. `dict()` : Fungsi ini digunakan untuk mengubah urutan tuple (kunci,nilai) menjadi **dictionary**.
10. `str()` : Digunakan untuk mengubah integer menjadi **string**.
11. `complex(real,imag)` : Fungsi ini mengubah bilangan real menjadi bilangan **complex(real,imag)**.
12. `chr(angka)`: Fungsi ini mengubah angka menjadi **karakter ASCII** yang sesuai.

Sumber: (https://www.geeksforgeeks.org/type-conversion-python/)

Contoh:
    Kita akan merubah tipe data input menjadi **integer**. Maka syntaxnya sebagai berikut:

```python
# print sum of two numbers taken from the user

num_1 = int(input("Enter first num: "))
num_2 = int(input("Enter second num: "))

print('Type of num_1:',type(num_1))
print('Type of num_2:',type(num_2))

result = num_1 + num_2
print("The sum of given numbers is : ", result)
```

Misalnya, pengguna memasukkan string untuk syntax dibawah ini:

```python
num = int(input('Enter age: '))

#Saat input di console, masukkan input 'umur' tanpa tanda kutip
```

## Beberapa nilai input dalam satu baris

Kita juga dapat meminta beberapa nilai secara langsung pada satu baris dengan hanya satu panggilan ke fungsi `input()`.
Misalnya, mari kita mendapatkan beberapa informasi tentang siswa dari pengguna dan menyimpannya dalam variabel yang berbeda.

```python
name, age, score = input("Enter student's name, age and score separated by space:").split()
print("Student Name:", name)
print("Student Age:", age)
print("Student Score:", score)

#Contoh input: Ghelby 14 88
```

Setelah memasukkan input maka output yang kita dapat adalah sebagai berikut:

    Student Name: Ghelby
    Student Age: 14
    Student Score: 88

Di sini, kita memisahkan string input dengan spasi menggunakan metode `split()`.

# Kondisi (Conditional)

## Kondisi If
Pengambilan keputusan (kondisi if) digunakan untuk mengantisipasi kondisi yang terjadi saat jalanya program dan menentukan tindakan apa yang akan diambil sesuai dengan kondisi.

Pada python ada beberapa statement/kondisi diantaranya adalah `if`, `else` dan `elif` Kondisi if digunakan untuk mengeksekusi kode jika kondisi bernilai benar `True`.

Jika kondisi bernilai salah `False` maka statement/kondisi `if` tidak akan di-eksekusi.

Dibawah ini adalah contoh penggunaan kondisi `if` pada Python

```python
#Kondisi if adalah kondisi yang akan dieksekusi oleh program jika bernilai benar atau TRUE

nilai = 9

#jika kondisi benar/TRUE maka program akan mengeksekusi perintah dibawahnya
if(nilai > 7): #ini adalah statement
    print("Selamat Anda Lulus") #ini adalah body

#jika kondisi salah/FALSE maka program tidak akan mengeksekusi perintah dibawahnya
if(nilai > 10):
    print("Selamat Anda Lulus")
```

Disana variabel `nilai` yaitu 9 sehingga yang dieksekusi hanya `if` pertama. Ganti nilai variabel `nilai` menjadi lebih dari 10 maka `if` kedua akan dieksekusi. Jika nilai variabel `nilai` kurang dari 7 maka kedua `if` tidak akan dieksekusi.

## Kondisi If Else

Pengambilan keputusan (kondisi if else) tidak hanya digunakan untuk menentukan tindakan apa yang akan diambil sesuai dengan kondisi, tetapi juga digunakan untuk menentukan tindakan apa yang akan diambil/dijalankan jika kondisi tidak sesuai.

Kondisi `if` `else` adalah kondisi dimana jika pernyataan benar `True` maka kode(body) dalam `if` akan dieksekusi, tetapi jika bernilai salah `False` maka akan mengeksekusi kode(body) di dalam `else`.

Dibawah ini adalah contoh penggunaan kondisi `if` `else` pada Python:

```python
#Kondisi if else adalah jika kondisi bernilai TRUE maka akan dieksekusi pada if, tetapi jika bernilai FALSE maka akan dieksekusi kode pada else

nilai = 3
#Jika pernyataan pada if bernilai TRUE maka if akan dieksekusi, tetapi jika FALSE kode pada else yang akan dieksekusi.
if(nilai > 7):
    print("Selamat Anda Lulus")
else:
    print("Maaf Anda Tidak Lulus")
```

Pada contoh diatas, jika program dijalankan maka akan mencetak string `"Maaf Anda Tidak Lulus"` karena pernyataan pada `if` bernilai `False`.

## Kondisi Elif
Pengambilan keputusan (kondisi if elif) merupakan lanjutan/percabangan logika dari "kondisi if". `elif` merupakan singkatan dari "else if". Dengan `elif` kita bisa membuat kode program yang akan menyeleksi beberapa kemungkinan yang bisa terjadi. Hampir sama dengan kondisi "else", bedanya kondisi "elif" bisa banyak dan tidak hanya satu.

Dibawah ini adalah contoh penggunaan kondisi elif pada Python:

```python
#Contoh penggunaan kondisi elif

hari_ini = "Minggu"

if(hari_ini == "Senin"):
    print("Saya akan kuliah")
elif(hari_ini == "Selasa"):
    print("Saya akan kuliah")
elif(hari_ini == "Rabu"):
    print("Saya akan kuliah")
elif(hari_ini == "Kamis"):
    print("Saya akan kuliah")
elif(hari_ini == "Jumat"):
    print("Saya akan kuliah")
elif(hari_ini == "Sabtu"):
    print("Saya akan kuliah")
elif(hari_ini == "Minggu"):
    print("Saya akan libur")
```

Pada contoh diatas, jika program dijalankan maka akan mencetak string "Saya akan libur".
