### Mendefinisikan Fungsi Python

Anda dapat menentukan fungsi untuk menyediakan fungsionalitas yang dibutuhkan. Berikut adalah aturan sederhana untuk mendefinisikan fungsi dengan Python.

- Fungsi blok dimulai dengan `def` kata kunci diikuti oleh **nama fungsi** dan tanda kurung (()). Contoh def my_function()
- Setiap parameter masukan atau argumen harus **ditempatkan** di dalam tanda kurung ini. Anda juga dapat menentukan parameter di dalam tanda kurung ini.
- Pernyataan pertama dari sebuah fungsi dapat berupa pernyataan opsional - string, dokumentasi, fungsi, atau docstring.
- Blok kode dalam setiap fungsi dimulai dengan titik dua (:) dan indentasi.
- Pernyataan `return` [ekspresi] keluar dari sebuah fungsi, secara opsional menyampaikan kembali ekspresi ke pemanggil. Pernyataan pengembalian tanpa argumen sama dengan return None.

Terdapat dua jenis functions di Python/
- *Built-in/pre-defined functions* yaitu fungsi yang sudah didefinisikan di Python. Contohnya `id()`, `type()`, dan `print()`.
- *User-Defined Functions/Customized Functions* yaitu fungsi yang didefinisikan oleh pengguna. Fungsi ini yang akan kita pelajari di pertemuan ini.

Syntax untuk mendefinisikan fungsi yaitu sebagai berikut:

    def function_name (Parameter) :
        “ Doc String “
        Body
        Return value

Catatan: `def` wajib ada sedangkan `return` opsional

Contoh mendefinisikan fungsi:

```python
#contoh 1
def nama() :
    print("Ghelby M Faid")
```

`Run` syntax tersebut sehingga fungsi `nama()` didefinisikan di Python. Setelah itu setiap kali kita memanggil `nama()`, maka `Ghelby M Faid` akan di print oleh Python.

```python
nama()
```

## Python Return Statement

Suatu fungsi dapat mengembalikan(**return**) beberapa nilai. Jadi, dengan logika yang kita berikan, kita akan mendapatkan beberapa hasil, dan hasil itu adalah **return**. Contoh:

```python
def penjumlahan(angka_1, angka_2):
    sum = angka_1 + angka_2
    return sum
```

Kita belum menanyakan hasil **sum**, tetapi Python tidak akan memberikan error karena tidak wajib mengambil nilai **return**.

Namun, kita tidak akan mendapatkan nilai apa pun karena kita belum memanggil fungsi **return**. Di sini, kita dapat mengeksekusinya dengan cara menyimpannya dalam sebuah variabel, dan kemudian mencetak hasilnya. Terdapat beberapa cara yang bisa digunakan contohnya:

```python
#cara 1
result = penjumlahan(1,2)
print(result)

#cara 2
print(penjumlahan(1,2))
```

Perlu diingat bahwa Python mengembalikan nilai **None** secara default.

Satu hal lagi, fungsi `return` yang telah kita bahas adalah opsional tetapi jika kita ingin menyimpan nilai dan menggunakannya lagi di luar fungsi, maka kita dapat menggunakannya.
